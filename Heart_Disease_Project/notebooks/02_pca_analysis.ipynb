## Dimensionality Reduction - PCA (Principal Component Analysis) 

```python
# Import needed library 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

```

```python
# Load Data 
X = pd.read_csv('../data/cleaned_X.csv')
y = pd.read_csv('../data/clean_y.csv')

# Scaling Data 
scaler = StandardScaler()
X = scaler.fit_transform(X)


```

```python
# Apply PCA 
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X)

# Plot Results 
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y['target'], cmap='viridis')
plt.xlabel('First Pricipal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Heart Disease Dataset')
plt.colorbar(label='target')
plt.show()
```

```python
# Calculate Cumulative Covariance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot the Cumulative Covariance 
plt.figure(figsize=(10, 8))
plt.plot((cumulative_variance),  'bo-')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.title('PCA: Cumulative Variance')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
